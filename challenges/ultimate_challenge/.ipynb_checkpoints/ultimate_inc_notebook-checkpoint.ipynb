### The weekly demand cycle:

Weekdays experience 2 peaks in demand. There is a rise at 9a from 3 users to 20 at 11:30a, then a decline to 5 users at 2p. There is a rise at 8p from 5 users to a variable peak at 11:30p, then a decline to a resting point of 3 users at 5a. Interestingly, the evening peak grows gradually through the weak from about 12 users on Monday night to about 30 on Friday night. What was suprising is the smooth increase in this peak's value, rather than a sudden jump at the weekend.

Between these peaks there are "low periods" from 5a to 9a and 2p to  8p. The afternoon rest period is a bit busier. On Fridays, the afternoon experiences a gradual rise from 5 to 10 riders between 5p and 8p on Fridays. 

Weekends are a different story. Logins sustain at the Friday night peak at 30 until about 5a on Saturday, when they sharply drop off to a handful of logins at 6am. This gradually picks up starting around noon to about 12 logins/hour at 2:30 and increases again to an evening peak at around 20 on 11:30p on Saturday. Note that the Saturday evening peak is lower than that of Friday evening. Sunday morning log ins gradually rise to about 35/hour at 10 am, and then decline to a handful in the morning, again increases to about 10 at noon. Notably, Sunday nights do not experience a demand peak, but do continue steadily around 10a.


It's clear that there is a weekly rhythm. It would be helpful to know if these logins are for riders or drivers, as this would paint a different picture. In the real world I would, of course, ask!

### Experiment Design

Next step:
* Do some background reading on power analysis.
* Build in reasonable guesswork for powering the study

Considerations:
* The metric will involve bridge crossings.
* Bridge crossings will usually happen in pairs - one to travel and one to go home.
* Bridge crossings will vary by time of day, which will also be reflected in the number of drivers on the road.
* Traffic should follow a daily consistent pattern on weekdays and a different pattern on weekends.
* Some drivers will take to the change, adopting the reiumbursement, and others will continue on as planned. What we care most about is this change in behavior.

Metric Design:
* We'll assign a feature to each driver called "is_two_city_driver"
* Two city drivers will be defined as crossing the bridge on at least x% of the days that they drive in the past two-weeks.
* An priori thought for the x threshold variable is 33%, exploratory analysis of the frequency of bridge crossings could help define exactly how often a two-city driver might tend to cross the bridge and how well this seperates them from one city drivers.

Experiment design:
* Each driver's two-city driver status is a boolean.

* By treating the before and after outcomes as realizations of a random variable, we can put a 90% confidence interval around the true percentage change in two-city drivership introduced by the experiment.

* We will use a McNemar's test to decide the statistical significance of the interventions effect on two-city drivership before and after the test.

* To account for drivers who quit using the app, we will sample larger than we need to begin with and take only a subset of drivers who remained during the full time period. This assumes that the bridge reimbursement doesn't cause people to quit- seems safe.

* Powering the study? Unknown.
https://egap.org/methods-guides/10-things-you-need-know-about-statistical-power

Recommendations:

* If a statistically significant change occurs and it is in the positive direction, keep doing what you're doing. If not, consider a different or complementary method for incentivizing the drivers.

### Simulation for Significance Test of Intervention

```python
#Glogal vars
np.random.seed(42)
n_drivers = 2500
# p_active_init = initial probability of being an active two-city driver
p_active_init = 0.1
# p_switch_up = probability that a currently non-active driver
# will be become active in test 2.
p_switch_up = 0.25
# p_switch_down = probability that a previously active driver 
# will stop crossing bridges in test 2.
p_switch_down = 0.05

def generate_data(n_drivers=n_drivers,
                   p_switch_up=p_switch_up,
                   p_switch_down=p_switch_down):
    init_activity = np.random.rand(n_drivers)
    init_activity = pd.Series(init_activity <= p_active_init).astype('int64')
    after_activity = []
    for i in init_activity:
        if i == 1:
            rand_gate = np.random.rand(1)[0] > p_switch_down
            if rand_gate:
                after_activity.append(1)
            else:
                after_activity.append(0)
        if i == 0:
            rand_gate = np.random.rand(1)[0] <= p_switch_up
            if rand_gate:
                after_activity.append(1)
            else:
                after_activity.append(0)
    after_activity = pd.Series(after_activity).astype('int64')
    df = pd.DataFrame(data={'before':init_activity,
                              'after':after_activity})
    df.index.name = 'user_num'
    return df
    
    
```

```python
df = generate_data()
```

```python
tab = pd.crosstab(df.before > 0,df.after > 0)
```

```python
print(mcnemar(tab,exact=True))
```

### Simulation for % Change Confidence Interval

```python
empirical_before = df.before.sum()/n_drivers
empirical_after = df.after.sum()/n_drivers
empirical_change = (empirical_after-empirical_before)
print("The observed change in two city drivership " +
    "was {}%, from {}% to {}%.".format(np.round(empirical_change*100,2),
                                    np.round(empirical_before*100,2),
                                    np.round(empirical_after*100,2)))
```

```python
# Treat each set of outcomes as a binomial random variable
n = n_drivers

p_before = df['before'].sum()/n_drivers
before_mean = n*p_before
before_var = n*p_before/(1-p_before)
p_before_conf = np.array([before_mean + 2*np.sqrt(before_var),
                         before_mean - 2*np.sqrt(before_var)])

p_after = df['after'].sum()/n_drivers
after_mean = n*p_after
after_var = n*p_after/(1-p_after)

# min effect = max_after - min_bfore, max effect = min_after - max_before.
p_after_conf = np.array([after_mean - 2*np.sqrt(after_var),
                         after_mean + 2*np.sqrt(after_var)])

effect_conf = (p_after_conf - p_before_conf)/n_drivers
```

```python
print("If repeated, the experiment would be 95% likely to give a result between " +
    "{}% and {}%.".format(np.round(effect_conf[0]*100,2),
                                    np.round(effect_conf[1]*100,2)))
```

# Predictive Model

```python
X = df.drop(['churned'],axis=1)
y = df['churned']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42, stratify=churned)
```

```python
lr_drops = ['avg_dist','avg_rating_of_driver','surge_pct']

X_train_lr = X_train.drop(lr_drops,axis=1)
X_test_lr = X_test.drop(lr_drops,axis=1)

clf = LogisticRegression()
clf.fit(X_train_lr,y_train)
y_pred = clf.predict(X_train_lr)
print(classification_report(y_train,y_pred))
```

Logistic regression has 70% precision and recall and tends to over-predict a bit in favor of "will churn", which, in this case, is probably better than underpredicting.


```python
rfc = RandomForestClassifier(max_features='sqrt')

gs = GridSearchCV(estimator=rfc,
                 param_grid={'n_estimators':[20,40,60,80,100]},
                 cv=5)
                  
gs.fit(X_train,y_train)
y_pred = gs.predict(X_train)
print(classification_report(y_train,y_pred))
```

Random Forest Manages to overfit a great deal it seems.

```python
print("Random Forest")
y_pred = gs.predict(X_test)
print(classification_report(y_test,y_pred))
print("Log Reg")
y_pred = clf.predict(X_test_lr)
print(classification_report(y_test,y_pred))
```

Random forest is the better model.
