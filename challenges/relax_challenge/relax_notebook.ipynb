### Creating Adopted User Feature

Calculate time interval between each login and the nth login after.
If any duration is less than or equal to 7 days, mark user as active.

```python
# This one takes a few seconds to run.
n = 2
adopted = []
for i in users.object_id.values:
    times = engage[engage['user_id'] == i]['time_stamp'].copy()
    starts = times[:-n].reset_index(drop=True)
    ends = times[n:].reset_index(drop=True)
    deltas = ends-starts
    adopted.append((deltas <= '7d').any())
users['adopted'] = adopted
users['adopted'] = users['adopted'].astype('int64')
```

### Categoricals -> Indicators, Exploratory Analysis

```python
users = users.rename(columns={'creation_source':'source',
                      'opted_in_to_mailing_list':'on_mailing_list',
                      'enabled_for_marketing_drip':'on_marketing_drip'
})
```

```python
X_tab = users[['domain','source','on_mailing_list','on_marketing_drip']]
X = pd.get_dummies(X_tab)
y = users['adopted']
Xy = pd.concat([X,y],axis=1)
```

### Predictive Modeling

Let's drop one of each one-hot categorical in order to avoid introducing redundant information in the model.

```python
X = X.drop(['domain_jourrapide.com','source_SIGNUP'],axis=1)
```

### Fix the class balancing.

```python
#Take a balanced sampling for an alternative training set and an unbalanced/stratified test set
np.random.seed(42)

class_balance = Xy.groupby('adopted')['on_mailing_list'].count().values
print(class_balance)

data_copy = Xy.copy()
training_set = np.array([])
test_set = np.array([])
smallest_class_total = np.min(class_balance)
test_frac = 0.15
training_size = int(np.floor(smallest_class_total*(1-test_frac)))
min_test_size = smallest_class_total-training_size
test_sizes = np.floor(class_balance*(min_test_size/smallest_class_total)).astype(int)

for counter, class_size in enumerate(class_balance):
    training_selections = np.random.choice(
        data_copy[data_copy['adopted'] == counter].index,
        size=training_size,
        replace=False
    )
    training_set = np.append(training_set,training_selections)
    data_copy = data_copy.drop(training_selections)
    
    test_selections = np.random.choice(
        data_copy[data_copy['adopted'] == counter].index,
        size=test_sizes[counter],
        replace=False
    )
    test_set = np.append(test_set,test_selections)
    data_copy = data_copy.drop(test_selections)

training_set = training_set.astype(int)
test_set = test_set.astype(int)
```

```python
X_train = X.loc[training_set]
y_train = y.loc[training_set]
X_test = X.loc[test_set]
y_test = y.loc[test_set]
```

```python
len(X_test)
```

### Due to lack of predictive power in covariates, naive bayes, log reg, and random forests are only a little better than a coin flip

```python
clf = BernoulliNB()
                  
clf.fit(X_train,y_train)
print("training set")
y_pred = clf.predict(X_train)
print(classification_report(y_train,y_pred))
print("test set")
y_pred = clf.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
probs = np.exp(clf.feature_log_prob_)
```

```python
p_diff = probs[1]-probs[0]
```
