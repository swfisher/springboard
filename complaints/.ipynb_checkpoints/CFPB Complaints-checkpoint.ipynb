# Extracting Value from Customer Complaints

Let's import some packages and implement some functions

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')

import re
import string

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

def dateLookup(s):
    """
    Memoization solution for faster date_time parsing
    source: https://stackoverflow.com/questions/29882573/pandas-slow-date-conversion
    """
    dates = {date:pd.to_datetime(date) for date in s.unique()}
    return s.map(dates)
```

```python
c = pd.read_csv('Consumer_Complaints.csv')

#Take subset where complaint text is available
c = c[c['Consumer consent provided?'] == 'Consent provided']

#Convert yes/no strings to bool
for col in ['Consumer disputed?','Timely response?']:
    c[col] = (c[col] == 'Yes')
    
#Convert datetimes    
for col in ['Date received', 'Date sent to company']:
    c[col] = dateLookup(c[col])
    
#All complaints in our subset are submitted via web and have consent provided.
c = c.drop(['Consumer consent provided?','Submitted via'],axis='columns')

#Convert all others, convert yes/no bools to int64
dataTypes = {
    'Product':'category',
    'Sub-product':'category',
    'Issue':'category',
    'Sub-issue':'category',
    'Consumer complaint narrative':str,
    'Company public response':str,
    'Company':'category',
    'State':'category',
    'ZIP code':str,
    'Tags':str,
    'Company response to consumer':'category',
    'Timely response?':'int64',
    'Consumer disputed?':'int64',
    'Complaint ID':'int64'
}

c = c.astype(dataTypes)

columnNames = {
    'Date received':'date_received',
    'Product':'product',
    'Sub-product':'sub_product',
    'Issue':'issue',
    'Sub-issue':'sub_issue',
    'Consumer complaint narrative':'text',
    'Company public response':'pub_reply',
    'Company':'company',
    'State':'state',
    'ZIP code':'zip_code',
    'Tags':'tags',
    'Date sent to company':'date_sent',
    'Company response to consumer':'cust_reply',
    'Timely response?':'timely_reply',
    'Consumer disputed?':'int64',
    'Complaint ID':'int64'
}

c = c.rename(columns = columnNames)

b = c[c['product'] == 'Bank account or service'].copy()

#There are a few missing state entries, and not every complaint has a sub-product or sub-issue.
```

### Cleaning Up for Word Vectorization.

```python
#Remove XXXXs
#Are numbers significant/helpful? Is there another format that might make them moreso?
#Strings of XXXX are used for redaction. These will need to be removed.
def remove_redact(text):
    return re.sub('XX/?','',text)

b.loc[:,'clean_text'] = b['text'].apply(remove_redact)
```

### Making a Random Sampling to Vectorize

```python
#Make the random sampling consistent when re-run.
np.random.seed(0)

#We'll take a sample of the whole collection and of each issue.
gross_sample_size = 1000;
issue_sample_size = 500;
issues = list(set(b.issue.unique()))

gross_text_sample = b.sample(gross_sample_size)['clean_text'].reset_index(drop=True)
issue_text_samples = {i:b[b['issue'] == i].sample(issue_sample_size)['clean_text'].reset_index(drop=True) for i in issues}
text_globs = pd.Series({i:' '.join(issue_text_samples[i]) for i in issues},index = issues)

print('There are ' + str(b.shape[0]) + ' entries.')
print('We\'ll use '+ str(gross_sample_size) + ' of them at random.')
print(issues)
```

### Create Bags of Words (Term Frequency) and Bag of Words/Inverse Document Frequency

The most frequent members of the bag of words for each issue show us that there seem to be unique and prevalent phrases for each issue. The Term Frequency/Inverse Document Frequency can confirm this. 

The phrases we're interested are phrases that are very commonly stated about a specific topic and rarely stated about other topics. These presence of these phrases would be predictive of the issue, especially as they co-occur and evidence accumulates. From a human perspective, our ticket to success will be phrases that act as a dead give away for the issue of the message. Conversely, we'll have plenty of garbage words which are common across tickets or appear too rarely to be helpful.

We can imagine these words as contributing evidence towards an indentifying label in a Bayesian sense. We can also imagine them as effectively "adding up to" a label in a regression model.

Term Frequency = Number of occurences of term within a "Document".

Inverse Document Frequency = 1 + Common Log of (Number of documents/(1+Number of documents containing word)) <- this is the fraction of documents containing the word inverted. We offset by 1 for DivByZero error safety.

TFIDF = Occurences in Document * Inverse Document Frequency. Grows with num occurence. Shrinks with portion of documents containing word


```python
def bow_vectorize(text_series):
    cv = CountVectorizer(stop_words='english',min_df=0.01, ngram_range=(2,30), max_features = 500)
    m = cv.fit_transform(text_series)
    return {'cv':cv,'m':m}

def tfidf(bow_m):
    tt = TfidfTransformer(norm='l2',smooth_idf=True,use_idf=True)
    m = tt.fit_transform(bow_m)
    return {'tt':tt,'m':m}
```

### How uniquely prevalent are certain phrases per issue?

We construct a bag of words matrix for (issue,word_frequency) and procedue to create (issue, word_tfidf).

```python
glob_bow = bow_vectorize(text_globs)
```

```python
tfidf_m = glob_tfidf['m'].todense().T
feature_names = glob_bow['cv'].get_feature_names()
tfidf_df = pd.DataFrame(tfidf_m, index=feature_names ,columns=issues)
tfidf_df_z = pd.DataFrame(stats.zscore(tfidf_df,axis=1),index=feature_names ,columns=issues)
```

```python
glob_tfidf = tfidf(glob_bow['m'])
```

These are the most uniquely prevalent words for each issue by tfidf z-score. Subjectively, they indeed seem to be phrases that are specific to the issue and commonly stated.

```python
for issue in issues:
    print(issue)
    print(tfidf_df[tfidf_df_z[issue] >= 1.8][issue])
    print()
```

```python
#Previous approach: Compute (ticket,frequency) bag of words matrix for issue and sum them.
issue_bag_of_words = {i: bow_vectorize(issue_text_samples[i]) for i in issues}
#Convert from scipy sparse csr_matrix to dense numpy matrix and get the word names
for issue in issues:
    m = issue_bag_of_words[issue]['m'].todense()
    col_names = issue_bag_of_words[issue]['cv'].get_feature_names()
    print(issue)
    sum_mat = m.sum(axis=0)
    issue_word_count = pd.DataFrame([sum_mat[0,i] for i in range(m.shape[1])],index=col_names,columns=['freq'])
    issue_word_count['z'] = stats.zscore(issue_word_count)
    plt.hist(issue_word_count['freq'])
    plt.show()
    print(issue_word_count[issue_word_count['z'] >= 1.0])
```

## Complaint Volume by State and Time

What's the average number of complaint per year for each state? Complaints per capita?
Get population data for 2015-2017 and update to reflect the average complaints per capita if interesting.

```python
fullyears = c['date_sent'] < '2018'

piv = c[fullyears].pivot_table(index=['state'],columns=c['date_sent'].dt.year,values='product'
                               ,aggfunc='count').fillna(0)

piv['mean'] = piv.mean(axis=1)
piv['std'] = piv.std(axis=1)
piv['zmean'] = stats.zscore(piv['mean'])
piv['zstd'] = stats.zscore(piv['std'])

piv.plot(kind='hist',y=['mean','std'],alpha=0.8,subplots=True,cmap='winter',title='Avg Complaints per Year by State')
plt.show()

#display outliers
print("Outlier states: ")
print(piv[['mean','std','zmean','zstd']][(np.abs(piv['zmean']) > 1.0) | (np.abs(piv['zstd']) > 1.0)])

piv2 = c.pivot_table(index = [c['date_sent'].dt.year, c['date_sent'].dt.month]
                                         ,values='product',aggfunc='count')
piv2.index.names = ['year','month']
piv2.columns = ['complaints per month']

print()
print("The number of complaints is generally increasing substantially.")
piv2.plot(kind='line',title = 'Total Complaints per Month', xlim=[0,33])
plt.show()
print("The ending drop off is odd. It seems likely that some data from november and december have not been added.")
print("It also appears that data collection may have been scarce at the beginning of the data set.")
print("2017 Monthly Complaints: ")
piv2.loc[(2017),:]

```
